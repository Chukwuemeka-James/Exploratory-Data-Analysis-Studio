### Let's loda the second dataset **Country-Code**

```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

### Let's merge the two dataframes

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
##To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

**Let's retrieve the list of unique country names from the Country column of the dataset, sorted by the frequency of their occurrence, and stores them as an index using `value_counts()`**

```python
country_names=final_df.Country.value_counts().index
country_names
```

**Here let's retrieve the corresponding frequency of occurrences (counts) of each country in the Country column and stores them as an array using the `.values` method.**

```python
country_val=final_df.Country.value_counts().values
country_val
```

### Let's draw a Pie Chart- Top 3 countries that uses zomato


```python
## Pie Chart- Top 3 countries that uses zomato

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

**Observation:Zomato maximum records or transaction are from India 
After that USA and then United Kingdoms**

```python
final_df.columns
```

**The code below groups the dataset by the rating, rating color, and rating text, then counts the number of occurrences for each combination and renames the resulting count column as 'Rating Count'.**

```python
the_ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
the_ratings.head()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When the Rating is between 4.5 to 4.9---> Excellent
2. When the Rating are between 4.0 to 4.4--->very good
3. when the Rating is between 3.5 to 3.9----> good
4. when the Rating is between 3.0 to 3.4----> average
5. when the Rating is between 2.5 to 2.9----> average
6. when the Rating is between 2.0 to 2.4----> Poor

Let's check the relationship between `Aggregate rating` and `Rating Count`

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show()
```

```python
# Bar plot
sns.barplot(
    x="Aggregate rating",
    y="Rating Count",
    hue='Rating color',
    data=ratings,
    palette=['blue', 'red', 'orange', 'yellow', 'green', 'green']
)

plt.show()
```

Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

### Let's replicate the above visualization with Countplot() which is usually used for categorical variables.

```python
sns.countplot(
    x="Rating color", 
    data=ratings,
    # Assign x variable to hue
    hue="Rating color",
    palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'], 
)
plt.show()
```
